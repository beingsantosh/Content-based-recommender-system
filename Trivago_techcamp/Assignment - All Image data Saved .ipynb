### Code for saving image name from several HTML files

1) There are several libraries used in coding like pandas, BeautifulSpoup, OS

2) Complete code is splited into small packages and run seperately to make it more understandable.



```python
# ImagesName to save in notepad file for all HTML files

import os
from bs4 import BeautifulSoup
basepath = os.getcwd()

note = open("all_images_allfiles.txt","w+")

for entry in os.listdir(basepath):
    if os.path.isfile(os.path.join(basepath, entry)):
        if entry.split('.')[-1] == 'html':
            fpath = f'{basepath}\{entry}'
            fcontents = open(fpath, 'r', encoding="utf8")
            soup = BeautifulSoup(fcontents, 'html.parser')
            note.write(fpath)
            note.write("\n" )
            for img in soup.find_all('img'):
                if (img.get('src')) is not None:
                    strm = f"{img.get('src')} {img.get('width')} {img.get('height')}"
                    note.write(strm)
                    note.write("\n" )
```

```python
# shift notepad file data in dataframe ( imagename)
import pandas as pd
df = pd.DataFrame(columns = ['HTML File Name','Image Source Name','Image width','Image Height'])
ite = 0
f = open("all_images_allfiles.txt", "r")
for st in f:
    if st.split(':')[0]=='C':
        htmlfilename = st.split('\\')[-1]
    else:
        st = st.split(' ')
        df.loc[ite] = [htmlfilename,st[0],st[1],st[2]]
        ite += 1
                                

# removing file extension
df['HTML File Name'] = df['HTML File Name'].apply(lambda item: item.split('.')[0])
# removing \n from end
df['Image Source Name']=df['Image Source Name'].apply(lambda item:item.split('\n')[0])
df['Image Height']=df['Image Height'].apply(lambda item:item.split('\n')[0])
# picking only file name and leaving all path
df['Image Source Name']=df['Image Source Name'].apply(lambda item:item.split('/')[-1])

df
```

```python
# saving df in csv
df.to_csv('imagename.csv')
```
