### Code for saving Links name from several HTML files inside a JSON file(and save it locally)

1) There are several libraries used in coding like pandas, BeautifulSpoup, OS, JSON

2) Complete code is splited into small packages and run seperately to make it more understandable.

3) Result will be in nested Dictionary format as below

data_dict = { 
               "HTML File Name" : {1:"FileName", 2:"Filename2",...... },                                 
               "Links Name"  : { 1: "link1", 2: "link2",.....}
             }                    



```python
# assignement 1
# links saved in notepad file for all HTML files

import os
from bs4 import BeautifulSoup
basepath = os.getcwd()
import json

note = open("all_links_allfiles.txt","w+")

for entry in os.listdir(basepath):
    if os.path.isfile(os.path.join(basepath, entry)):
        # pick only html files - if condition to check 
        if entry.split('.')[-1] == 'html':
            fpath = f'{basepath}\{entry}'     # complete path 
            fcontents = open(fpath, 'r', encoding="utf8")     #reading the content in html
            soup = BeautifulSoup(fcontents, 'html.parser')    
            note.write(fpath)                                  #write down filepath in notepad file    
            note.write("\n" )
            for link in soup.find_all('a'):                    # look for only links 
                if (link.get('href')) is not None:    
                    note.write(link.get('href') )
                    note.write("\n" )                       # write down all link name to notepad file

# all links saved at base location in 'all_links_allfiles.txt' file
```

```python
# Making dataframe to visualize it of all the links with its corresonding html file
import pandas as pd
df = pd.DataFrame(columns = ['HTML File Name','Links Name'])
# df.loc[4] = ['assas','wewewe']
ite = 0
f = open("all_links_allfiles.txt", "r")
for st in f:
#    If condition to check for file path like 'C:\wikipedia'
    if st.split(':')[0]=='C':
        htmlfilename = st.split('\\')[-1]
    else:
        df.loc[ite] = [htmlfilename,st]
        ite += 1
                                
df.head()
```

```python
# removing the file extension '.html' from HTML File Name column
df['HTML File Name'] = df['HTML File Name'].apply(lambda item: item.split('.')[0])
# removing \n from end of Links Name column
df['Links Name'] = df['Links Name'].apply(lambda item: item.split('\n')[0])
df.head()
```

```python
# make a dictionary dataset from df
data_dict = df.to_dict()
data_dict
```

```python
# make json dataset from dictionary 
jsonfile = json.dumps(data_dict,indent=4)
jsonfile
```

```python
#to save json file locally , please check in the folder with "datafinal.json" name
with open('datafinal.json', 'w') as f:
    json.dump(data_dict, f)
```
